# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# Extracting Year

```python
date["date_year"] = date['date'].dt.year
date.head()
```

# Extracting Month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

# Extracting Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# Day of the week 
date['date_day_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# Day of week -- name
date['date_day_of_week_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend ??
date['date_is_weekend'] = np.where(date['date_day_of_week_name'].isin(["Sunday",'Saturday']),1,0)
date.head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.week #  Please use Series.dt.isocalendar().week instead. 
date.head()
```

# Extract Quater

```python
date['quarter'] =  date['date'].dt.quarter
date.head()
```

# Extracting Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```

# Exact time elapsed between dates

```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```

# Time difference

```python
# in seconds

(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
